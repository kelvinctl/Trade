```python
import pandas_datareader as web
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 
import random
```

```python
ticker1 = 'QQQ'
ticker2 = 'KO'
ticker3 = 'EQIX'
ticker4 = 'TSM'
ticker5 = 'NFLX'
end = dt.datetime(2008, 12, 31)
#end = dt.datetime.now() 
start = end - dt.timedelta(days=3650)
days = 10
days1 = 100

data1 = web.DataReader(ticker1, 'yahoo', start, end)
data2 = web.DataReader(ticker2, 'yahoo', start, end)
data3 = web.DataReader(ticker3, 'yahoo', start, end)
data4 = web.DataReader(ticker4, 'yahoo', start, end)
data5 = web.DataReader(ticker5, 'yahoo', start, end)
```

```python
delta1 = data1['Adj Close'].diff(1)
delta1.dropna(inplace=True)

price1 = data1['Adj Close']
positive1 = delta1.copy()
negative1 = delta1.copy()

positive1[positive1 < 0] = 0
negative1[negative1 > 0] = 0


average_gain1 = positive1.rolling(window=days).mean()
average_loss1 = abs(negative1.rolling(window=days).mean())
SMA1_100 = price1.rolling(window=days1).mean()
relative_strength1 = average_gain1 / average_loss1
RSI1 = 100.0 - (100.0 / (1.0 + relative_strength1))

combined1 = pd.DataFrame()
combined1['Adj Close'] = data1['Adj Close']
combined1['RSI'] = RSI1
combined1['Name'] = ticker1
combined1['SMA100'] = SMA1_100
combined1['Price - SMA100'] = combined1['Adj Close'] - SMA1_100
combined1.dropna(inplace=True)
combined_for_analysis1 = combined1.loc[(combined1['RSI'] <= 30) | (combined1['RSI'] >= 70)] 
combined_for_analysis1.to_csv('test1.csv')

combined_for_analysis1['whole'] = combined_for_analysis1['Adj Close'].astype(str) + ',' + combined_for_analysis1['RSI'].astype(str) + ',' + combined_for_analysis1['Name'] + ',' + combined_for_analysis1['SMA100'].astype(str)
```
